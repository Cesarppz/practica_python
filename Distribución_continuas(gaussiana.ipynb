```
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import  norm
```

```
# definimos nuestra distribución gaussiana
def gaussian(x, mu, sigma):
 return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*pow((x-mu)/sigma,2))
```

```
x = np.arange(-4,4,0.1)
y = gaussian(x, 0.0 , 1.0)

plt.plot(x,y)
```

```
dist = norm(0 , 1) # Una distribución con promedio 0 y desviación estándar 1 
x = np.arange(-4 , 4 , 0.1)
y = [dist.pdf(values) for values in x]
                                                                                 # Distribucion de densidad que se calcula con el pdf
plt.plot(x,y)
```

```
dist = norm(0,1)
x = np.arange(-4,4,0.1)                                                         # Distribución de probabilidad acumulada 
y = [dist.cdf(values) for values in x]

plt.plot(x,y,color="red")
```

```
df = pd.read_excel("s057.xls") # Leer el archico exel en pandas
arr = df['Normally Distributed Housefly Wing Lengths'].values[4:]      #Busco los datos de cierta columna y específicamente los valores desde el indice 4
values, dist=np.unique(arr, return_counts=True)  #Busco cuales son los valores que se repiten, y me devuelve los nombres de los valores con su respectiva frecuencia  

plt.bar(values,dist)
```

```
# # estimación de la distribución de probabilidad
mu = arr.mean()                                                                 #promedio del arreglo
sigma = arr.std()                                                               #Desviación estándar del arreglo

# Distribución teórica                                                          
x = np.arange(30,60,0.1)
dist = norm(mu,sigma)
y=[dist.pdf(value) for value in x]
plt.plot(x,y,color="red")
plt.xlabel('valores')
plt.ylabel('probabilidades')
#datos
values, dist=np.unique(arr, return_counts=True) 
plt.bar(values,dist/len(arr))
```
